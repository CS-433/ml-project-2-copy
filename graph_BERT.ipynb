```python
lambda_list = ['2e-5', '3e-5','5e-5']
```

```python
accs = [[acc_b_custom_2_epoch, acc_b_class_2_epoch], [acc_b_custom_3_epoch, acc_b_class_3_epoch]]
with sns.axes_style("whitegrid"):
    fig, axs = plt.subplots(1,2,figsize=(15,6))

c = ['red', 'orange', 'teal']
for i in range(2):
    for j, lr in enumerate(lambda_list):
        if(i):
            n_epochs = [1, 2, 3]
        else: 
            n_epochs = [1, 2]
        accuracies=accs[i]
        #plot for custom
        axs[i].plot(n_epochs, accuracies[0][j], label="λ = {lr}, {m} class.".format(lr=lr, m="custom"), linewidth=3,color=c[j])
        #plot for classification
        axs[i].plot(n_epochs, accuracies[1][j], label="λ = {lr}, {m} class.".format(lr=lr, m="default"), linewidth=3, linestyle='--',color=c[j])
        #axs[i].fill_between(deg_list, accuracy[j,:]-std[j,:], accuracy[j,:]+std[j,:],alpha=0.10,color=c[j])
        axs[i].set_title("Grid Search on BERT, trained on {} epochs ".format(n_epochs[-1]),fontsize=20)
        axs[i].set_xlabel("Epoch",fontsize=20)
        axs[i].set_ylabel("Accuracy [%]",fontsize=20)
        axs[i].legend(loc='upper left')
        axs[i].set_ylim([87, 89])
        axs[i].set_xlim([1-0.05, n_epochs[-1]+0.05])
        axs[i].tick_params(axis='both', which='major', labelsize=20)
        axs[i].set_xticks(n_epochs)
fig.tight_layout()
fig.savefig('gridsearch_BERT.jpg', dpi=300)
```

```python
def vis_grid_search_acc(accs, lambda_list = ['2e-5', '3e-5','5e-5'], save=False):
    """
    Plot accuracy in function of the degree for different lambda values for every model
    """
    with sns.axes_style("whitegrid"):
        fig, axs = plt.subplots(1,2,figsize=(15,6))

    c = ['red', 'orange', 'teal']
    for i in range(2):
        for j, lr in enumerate(lambda_list):
            if(i):
                n_epochs = [1, 2, 3]
            else: 
                n_epochs = [1, 2]
            accuracies=accs[i]
            #plot for custom
            axs[i].plot(n_epochs, accuracies[0][j], label="λ = {lr}, {m} class.".format(lr=lr, m="custom"), linewidth=3,color=c[j])
            #plot for classification
            axs[i].plot(n_epochs, accuracies[1][j], label="λ = {lr}, {m} class.".format(lr=lr, m="default"), linewidth=3, linestyle='--',color=c[j])
            #axs[i].fill_between(deg_list, accuracy[j,:]-std[j,:], accuracy[j,:]+std[j,:],alpha=0.10,color=c[j])
            axs[i].set_title("Grid Search on BERT, trained on {} epochs ".format(n_epochs[-1]),fontsize=20)
            axs[i].set_xlabel("Epoch",fontsize=20)
            axs[i].set_ylabel("Accuracy [%]",fontsize=20)
            axs[i].legend(loc='upper left')
            axs[i].set_ylim([87, 89])
            axs[i].set_xlim([1-0.05, n_epochs[-1]+0.05])
            axs[i].tick_params(axis='both', which='major', labelsize=20)
            axs[i].set_xticks(n_epochs)
            
    fig.tight_layout()
    if(save):
        fig.savefig('gridsearch_BERT.jpg', dpi=300)
```

```python
accs = [[acc_b_custom_2_epoch, acc_b_class_2_epoch], [acc_b_custom_3_epoch, acc_b_class_3_epoch]]

vis_grid_search_acc(accs= accs)
```
